```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('LoveHina_vol14.xml')
root = tree.getroot()
root
```

```python
pages_list = []

for neighbor in root.iter('page'):
    page_dict = {}
    print(neighbor.attrib)
    page_dict['attrib'] = neighbor.attrib
    page_dict['frames'] = []
    for frame in neighbor.iter('frame'):
        print(frame.attrib)
        print(frame.attrib['id'])
        page_dict['frames'].append(frame.attrib)

    page_dict['lines'] = []
    for text in neighbor.iter('text'):
        print(text.attrib)
        print(text.attrib['id'])
        print(text.text)
        line_dict = {}
        line_dict['attrib'] = text.attrib
        line_dict['text_jp'] = text.text
        line_dict['text_pl'] = ""
        line_dict['read_order'] = 0

        page_dict['lines'].append(line_dict)

    pages_list.append(page_dict)


```

```python
pages_list
```

```python
import json

final = json.dumps(pages_list, indent=2, ensure_ascii=False)
```

```python
print(final)
```

```python
with open('LoveHina_vol14.json', 'w', encoding='utf-8') as f:
    json.dump(pages_list, f, ensure_ascii=False, indent=2)
```
